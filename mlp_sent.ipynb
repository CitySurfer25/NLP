```python
import pandas as pd
import numpy as np
df = pd.read_excel("Bank.xlsx")
```

```python
df
```

```python
df=df.drop(['Unnamed: 0'],axis=1)
df=df.drop([93,94,95,96,97,98,99,100],axis=0)
```

```python
df[['Sentences for Training ','Sense of the word "Bank"']]=df[['Sentences for Training ','Sense of the word "Bank"']].astype("string")
```

```python
df.dtypes
```

```python
df['Sense of the word "Bank"'].unique()
```

```python
import nltk
```

```python
dic = {}
column = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v',
          'w','x','y','z','label']
for i in column:
    dic[i] = []
    
```

```python
for sent in df['Sentences for Training ']:
    for i in column[:-1]:
        if i in sent:
            dic[i].append(sent.count(i))

        else:
            dic[i].append(0)
    dic["label"].append('')
```

```python
dic
```

```python
df_fin = pd.DataFrame(dic)
df_fin
```

```python
df_fin['label'] = df['Sense of the word "Bank"']
```

```python
df_fin
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
X = df_fin.drop(columns = ['label'])
y = df_fin['label']
```

```python
y = LabelEncoder().fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,y, test_size=0.1, random_state=42)
```

```python
model = MLPClassifier(hidden_layer_sizes = (100,50,25), activation='relu', max_iter = 25, verbose=1)
```

```python
model.fit(xtrain,ytrain)
y_pred = model.predict(xtest)
accuracy = accuracy_score(ytest, y_pred)*100
print("Accuracy :",accuracy)
```
