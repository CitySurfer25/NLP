# T - test calculation 

```python
#To calculate bigram values for the given text file and store that in dictionary for future access
def bigram_calc(word_tokens):
    bigrams = [(s1, s2) for s1, s2 in zip(word_tokens, word_tokens[1:])]
    d={}
    for (ww1,ww2) in set(bigrams):
        bg_count = bigrams.count((ww1,ww2))
        d[(ww1,ww2)]=bg_count/len(bigrams) 
    return d
```

```python
#To access bigram value for the particular pair in the bigram dictionary
def get_bigram_value(w1,w2,word_tokens):
    dic=bigram_calc(word_tokens)
    return dic[(w1,w2)]
```

```python
#Calculate T-test
def t_test_calc(w1,w2,word_tokens):
    w1_count=word_tokens.count(w1)
    w2_count=word_tokens.count(w2)
    p_w1=(w1_count/len(word_tokens))
    p_w2=(w2_count/len(word_tokens))
    expected_mean=p_w1*p_w2
    observed_mean=get_bigram_value(w1,w2,word_tokens)
    sample_var=observed_mean*(1-observed_mean)
    t=(observed_mean-expected_mean)/math.sqrt((sample_var/len(word_tokens)))
    return t
```
