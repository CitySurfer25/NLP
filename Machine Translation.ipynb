```python
import tensorflow as tf
from keras.utils import to_categorical
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding,Dense,SimpleRNN
import numpy as np
import pandas as pd
```

```python
df=pd.read_table("tam.txt",header=None)
```

```python
df
```

```python
source_text=df[0]
target_text=df[1]
```

```python
source_vocab=set(" ".join(source_text))
target_vocab=set(" ".join(target_text))
source_vocab_size=len(source_vocab)
target_vocab_size=len(target_vocab)
```

```python
source_ctoi={char:idx for idx,char in enumerate(source_vocab)}
target_ctoi={char:idx for idx,char in enumerate(target_vocab)}
source_itoc={idx:char for char,idx in source_ctoi.items()}
target_itoc={idx:char for char,idx in target_ctoi.items()}
```

```python
source_sequence= [[source_ctoi[char] for char in text]for text in source_text]
target_sequence= [[target_ctoi[char] for char in text]for text in target_text]
```

```python
max_length_sequence=max(len(seq)for seq in source_sequence)
source_sequence=pad_sequences(source_sequence,maxlen=max_length_sequence,padding='post')
target_sequence=pad_sequences(target_sequence,maxlen=max_length_sequence,padding='post')
```

```python
target_sequence_onehot=np.array([to_categorical(seq,num_classes=target_vocab_size)for seq in target_sequence])
```

```python
input_shape=(max_length_sequence,source_vocab_size)
output_shape=(max_length_sequence,target_vocab_size)
model=Sequential()
model.add(Embedding(source_vocab_size,64))
model.add(SimpleRNN(128,return_sequences=True))
model.add(Dense(target_vocab_size,activation='softmax'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(source_sequence,target_sequence_onehot,epochs=2000)
```

```python
input_sequence="I'll walk."
input_sequence=[source_ctoi[char]for char in input_sequence]
input_sequence=pad_sequences([input_sequence],maxlen=max_length_sequence,padding='post')
output_sequence=model.predict(input_sequence)[0]
```

```python
output_sequence=[target_itoc[np.argmax(char)]for char in output_sequence]
```

```python
print("Input Sequence: I'll walk.")
print("Translated Sequence:", ''.join(output_sequence))
```
